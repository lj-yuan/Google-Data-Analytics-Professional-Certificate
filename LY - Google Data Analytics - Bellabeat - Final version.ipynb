### Data preprocessing 


Deleting the duplicates in the sleep dataset

````python
sleep_df = sleep_df.drop_duplicates()

# this will show the result of our manipulation
# if 0, then there are no duplicates 
sleep_df.duplicated().sum() 
````

Converting 'ActivityDate' and 'SleepDay' to `datatime64` dtype

````python
daily_activity["ActivityDate"] = pd.to_datetime(daily_activity["ActivityDate"], format="%m/%d/%Y")
````

````python
pd.options.mode.chained_assignment = None  # hides the warning
sleep_df["SleepDay"] = pd.to_datetime(sleep_df["SleepDay"])
````

Creating new column by separating the date into day of the week for further analysis.

````python
# create new column "DayOfTheWeek" to represent day of the week 
daily_activity["DayOfTheWeek"] = daily_activity["ActivityDate"].dt.day_name()

# print 1st 5 rows to confirm
daily_activity["DayOfTheWeek"].head(7)
````

````python
# grouping total steps amount by days of the week and creating a new dataframe for it
total_steps = daily_activity \
        .groupby(['DayOfTheWeek'], as_index = False) \
        .agg({'TotalSteps': 'sum'})
total_steps
````

Creating new column *TotalMins* being the sum of total time logged.

````python
# create new column "TotalMins" containing sum of total minutes.
daily_activity["TotalMins"] = daily_activity["VeryActiveMinutes"] + daily_activity["FairlyActiveMinutes"] + daily_activity["LightlyActiveMinutes"] + daily_activity["SedentaryMinutes"]

daily_activity["TotalMins"].head(5)
````

Creating new column by converting *TotalHours*  to number of hours - TotalHours.

````python
# create new column "TotalHours" by converting to hour and round float to two decimal places
daily_activity["TotalHours"] = round(daily_activity["TotalMins"] / 60)

# print 1st 5 rows to confirm
daily_activity["TotalHours"].head(5)
````

````python
# create new column "DayOfTheWeek" to represent day of the week for sleep dataset 
sleep_df['DayOfTheWeek'] = (sleep_df.SleepDay.dt.day_name())
sleep_df.head()
````

````python
sleep_df['TotalHoursAsleep'] = round(sleep_df['TotalMinutesAsleep'] / 60, 2)
sleep_df.head()
````

##### **10,000 steps goal:** 

To better understand the daily activity levels of our users, I will calculate the average percentage of users who have reached the recommended 10,000 steps goal. Analyzing how many users on average are hitting this daily step target will provide insight into the overall activity patterns across our user base.

````python
# 10 000 steps is a common daily goal for an adult person
# creating a column to evaluate the goal success rate
daily_activity['StepsGoalReached'] = daily_activity['TotalSteps'] * 100 / 10000
````

````python
# creating a dataframe to find out an average goal success rate per user 
mean_goal_by_id = daily_activity \
                    .groupby('Id', as_index = False) \
                    .agg({'StepsGoalReached': 'mean'}) \
                    .sort_values('StepsGoalReached')

mean_goal_by_id.head()
````

````python
import seaborn as sns
import matplotlib.pyplot as plt

# making a barplot to visualize findings

ax = sns.barplot(
    x='Id',
    y='StepsGoalReached',
    data=mean_goal_by_id,
    order=mean_goal_by_id['Id'],
    color='#53777a',
    linewidth=2
)

# customizing the plot

plt.title('How many users have reached the 10 000 steps goal', fontsize=22)
plt.xlabel('Users', fontsize=16)
plt.ylabel('The percentage of average goal success rate', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(range(0, 110, 10))
plt.axhline(y=100, color='#f44336', linestyle='--')
plt.rcParams["figure.figsize"] = (17, 8)
plt.rcParams['font.size'] = 12

plt.show()
````

The chart above reveals that most FitBit users are not reaching the daily 10,000 step goal, with only 21% of users (7 out of 33) recording an average of 10,000+ steps.

##### **Calories Burned vs. Steps:**

````python
# Imports
import matplotlib.pyplot as plt
import seaborn as sns

# Use seaborn style defaults 
sns.set_style('darkgrid')

# Create figure and axes
fig, ax = plt.subplots(figsize=(10,6))

# Scatter plot with color mapping
sns.scatterplot(data=daily_activity, 
                x='TotalSteps', 
                y='Calories',
                hue='Calories',
                palette='coolwarm',
                alpha=0.8,
                ax=ax)

median_calories = 2304
median_steps = 7638

# Add median lines
ax.axvline(median_steps, color='black', ls='--', label='Average steps count (7638)') 
ax.axhline(median_calories, color='red', ls='--', label='Average calories count (2304)')

# Labels and titles
ax.set_xlabel('Total Steps', size=14)
ax.set_ylabel('Calories Burned', size=14)
ax.set_title('Calories Burned vs. Steps', size=18)

# Legend
ax.legend(fontsize='12')

# Tight layout
plt.rcParams['figure.figsize'] = [14, 8]
plt.show()
````

As we can see from the scatter plot above:

- The scatter plot shows a positive correlation between number of steps taken and calories burned, indicating that more steps generally leads to more calories burned.
- There is one outlier data point showing over 35,000 steps with less than 3,000 calories burned. This outlier could be explained by an unusually high step count for that user on that particular day or potentially an error in the data collection.
- Some users with zero steps recorded still burned up to 2,000 calories. Meanwhile, other users who were active and walked more burned a similar number of calories. This suggests that the difference between 'Very Active' and 'Fairly Active' minutes in users' daily activity, beyond just steps, likely explains variations in calories burned for given step counts.

##### **Calories Burned vs. Activity Hours**

````python
# Imports 
import matplotlib.pyplot as plt
import seaborn as sns

# Use seaborn styling 
sns.set_style('darkgrid') 

# Create figure and axes
fig, ax = plt.subplots(figsize=(10,6))

# Scatterplot 
sns.scatterplot(data=daily_activity,
                x='TotalHours', 
                y='Calories',
                hue='Calories',
                palette='RdYlGn',
                alpha=0.8,
                ax=ax)

median_calories = 2304
median_hours = 20
median_sedentary = 991 / 60

# Add median lines
ax.axvline(median_hours, c='blue', ls='--', label='Median Hours')
ax.axvline(median_sedentary, c='purple', ls='--', label='Median Sedentary Hours')  
ax.axhline(median_calories, c='red', ls='--', label='Median Calories')

# Axis labels  
ax.set_xlabel('Total Hours Logged', size=14)
ax.set_ylabel('Calories Burned', size=14)

# Title
ax.set_title('Calories Burned vs. Activity Hours', size=18)  

# Legend
ax.legend(loc='upper left', frameon=False)

# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Tight layout
plt.tight_layout()
plt.show()
````

   
The scatter plot shows a mild positive correlation between logged hours and calories burned, indicating that longer logged hours do not consistently translate to proportionally higher calories burned. This is evidenced by the cluster of average sedentary hours (purple line) around 16-17 hours.

There are also a few outliers in the data:
- Multiple data points with zero values for both logged hours and calories, possibly from tracking errors or non-representative days.
- One uncommon data point at 24 logged hours and zero calories burned, which could stem from the same issues as the zero-value outliers.

##### **Breakdown of Minutes for Different Activity Levels**


````python
import matplotlib.pyplot as plt
import numpy as np

# calculating total of individual minutes column
very_active_mins = daily_activity["VeryActiveMinutes"].sum()
fairly_active_mins = daily_activity["FairlyActiveMinutes"].sum()
lightly_active_mins = daily_activity["LightlyActiveMinutes"].sum()
sedentary_mins = daily_activity["SedentaryMinutes"].sum()

# Data and labels
slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very Active", "Fairly Active", "Lightly Active", "Sedentary"]
colors = ["#FF9999", "#66B3FF", "#99FF99", "#FFCC99"]  # Softer pastel colors
explode = [0.1, 0, 0, 0.1]

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting the pie chart
wedges, texts, autotexts = ax.pie(slices, labels=labels, colors=colors,
                                   autopct="%1.1f%%", startangle=140,
                                   explode=explode, shadow=True)

# Customize text properties
for text in texts:
    text.set_fontsize(12)
for autotext in autotexts:
    autotext.set_fontsize(12)

# Equal aspect ratio ensures that pie is drawn as a circle
ax.axis("equal")

# Adding a title
plt.title("Breakdown of Minutes for Different Activity Levels", fontsize=16, fontweight="bold")

# Display the plot
plt.tight_layout()
plt.show()
````


Observing the pie chart reveals the following insights:

1. The largest portion, accounting for 81.3%, is attributed to Sedentary minutes.

2. This suggests that users primarily utilize the FitBit app to record routine activities like commuting, passive movements, or errands. 

3. There is a minimal representation of fitness-related activities, as evidenced by the marginal percentages of fairly active activity (1.1%) and very active activity (1.7%). This trend is concerning given that the primary purpose of the FitBit app is to promote fitness engagement.
